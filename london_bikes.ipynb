```python
new_cols_dict = {
    'timestamp' : 'time',
    'cnt' : 'count',
    't1' : 'temp_actual_F',
    't2' : 'temp_feel_like_F',
    'hum' : 'humidity_percentage',
    'wind_speed': 'wind_speed_mph',
    'weather_code' : 'weather',
    'is_holiday' : 'is_holiday',
    'is_weekend': 'is_weekend',
    'season': 'season'
}

bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
# convert % to decimal
bikes.humidity_percentage = bikes.humidity_percentage / 100

# convert to US/Imperial units
bikes.temp_actual_F = bikes.temp_actual_F * 1.8 + 32
bikes.temp_feel_like_F = bikes.temp_feel_like_F * 1.8 + 32
bikes.wind_speed_mph = round(bikes.wind_speed_mph * 0.6213711922, 1)
```

```python
bikes
```

```python
season_dict = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'fall',
    '3.0': 'winter'
}


weather_dict = {
    '1.0' : 'Clear',
    '2.0' : 'Scattered Clouds',
    '3.0' : 'Partly Cloudy',
    '4.0' : 'Cloudy',
    '7.0' : 'Rainy',
    '10.0': 'Thunderstorms',
    '26.0': 'Snowy'
}


```

```python
bikes.season = bikes.season.astype('str')
bikes.season = bikes.season.map(season_dict)

bikes.weather = bikes.weather.astype('str')
bikes.weather = bikes.weather.map(weather_dict)

bikes
```

```python
bikes.to_excel('london_bikes_final.xlsx', sheet_name='Data')
```
